**SORTING A LIST**

> the sort() method changes the order of a list parmanently. the sorted() function 
returns a copy of the list, leaving the orginal list unchanged you can sort the items in alist in albhabethical order, or reverse alphabethical order. you can also reverse the original order of the list keep in mind that lowercase and upprcase latters may affect the sort order.



```python
users = ['lili', 'dodo','chini', 'toto','tintin']
print(users)
users.sort()
print(users.sort())
print(users)    #if i am using users.sort() or any opeartions on the process
# then i have to use print on the original operations. because we are performing operations on the orginal form
users.sort()  #sorting parmanently
print(users)
# SORTING A LIST PARMANENTLY IN REVERSE ALPHABETICAL ORDER
users.sort(reverse=True)
print(users)

#sorting a list temporarily
print(sorted(users))
print(sorted(users, reverse=True))
users.reverse()
print(users)
```

 **USING A LOOP TO GENERATE A LIST OF SQUARE NUMBERS**

```python
squares = []
for x in range(1,11):
  square = x**2
  squares.append(square)
  print(squares)
squares = [x**2 for x in range(1,11)]
print(squares)
squares = []
for x in squares:
  squares = [x**2 for x in range(1,11)]
  print(squares)

squares=[]
for x in range(1,11):
  square = x**2
  squares.append(square)
  print(squares)
```
